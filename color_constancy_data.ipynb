```python
import utils.utils_data as utils_data
import utils.color_constancy as color_constancy
import matplotlib.pyplot as plt
import os
from tqdm import tqdm_notebook as tqdm
import cv2
import numpy as np
```

## Shaded of gray

```python
for train_file in tqdm(utils_data.TRAIN_IMGS):
    new_file = train_file.replace("Train", "shade_of_gray_train")
    
    if not os.path.exists("/".join(new_file.split("/")[:-1])):
        os.makedirs("/".join(new_file.split("/")[:-1]))
        
    image_np = cv2.imread(train_file)
    image_np = cv2.cvtColor(image_np, cv2.COLOR_BGR2RGB) # Take care, plot some imgs and see results
    
    new_image_np = color_constancy.shade_of_gray(image_np, power=6, gamma=2.2)
    
    cv2.imwrite(new_file, cv2.cvtColor(new_image_np, cv2.COLOR_RGB2BGR))
```

```python
for val_file in tqdm(utils_data.VALIDATION_IMGS):
    new_file = val_file.replace("Train", "shade_of_gray_train")
    
    if not os.path.exists("/".join(new_file.split("/")[:-1])):
        os.makedirs("/".join(new_file.split("/")[:-1]))
        
    image_np = cv2.imread(val_file)
    image_np = cv2.cvtColor(image_np, cv2.COLOR_BGR2RGB) # Take care, plot some imgs and see results
    
    new_image_np = color_constancy.shade_of_gray(image_np, power=6, gamma=2.2)
    
    cv2.imwrite(new_file, cv2.cvtColor(new_image_np, cv2.COLOR_RGB2BGR))
```
